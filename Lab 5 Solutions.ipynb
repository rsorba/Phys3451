```python
import numpy as np # linear algebra
from scipy import special
import matplotlib.pyplot as plt
```

```python
# define the variables
a0 = 0.529e-11
r = np.linspace(0,25,10000)*a0
theta = np.linspace(0,2*np.pi,1000)
phi = np.linspace(0,2*np.pi,1000)
# and constants
mp = 1.672e-27 # proton mass in kg
hbar = 1.055e-34 # joule-sec
```

```python
def radial(n,ell,r):

    E = -13.6 * 1.602e-19/n**2 # energy in joules
    rho = np.sqrt(-2 * mp * E/hbar**2)*r
    #print(np.exp(-rho))
    Prad = np.exp(-rho) * rho **(2* ell) * (special.assoc_laguerre(rho,n,ell))**2*(4 * np.pi*r**2)
    return Prad
```

```python
plt.figure()
n = 1
ell = 0
plt.plot(r/a0,radial(n,ell,r))
```

```python
plt.figure()
n = 2
ell = 1
plt.plot(r/a0,radial(n,ell,r))
```

```python
plt.figure()
n = 3
ell = 1
plt.plot(r/a0,radial(n,ell,r))
```

```python
def thetaComponent(ell,ml,theta):
    Ptheta = np.sin(theta)**ml * special.eval_legendre(ell,np.cos(theta))
    return Ptheta**2
```

```python
ell = 3
ml = 0
for ml in [0,1,2,3]:
    fig = plt.figure()
    ax = fig.add_subplot(111,polar=True)
    ax.set_theta_zero_location("N")
    plt.plot(theta,thetaComponent(ell,ml,theta))
```
